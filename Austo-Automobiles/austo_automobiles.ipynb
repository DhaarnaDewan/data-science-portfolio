```python
# What is the likelihood of a salaried person buying a sedan?
```

```python
# Ensure consistent capitalization in Profession and Make columns
df['Profession'] = df['Profession'].str.lower()
df['Make'] = df['Make'].str.lower()

# 1. Filter for Salaried People
salaried_df = df[df['Profession'] == 'salaried']

# 2. Count Sedan Buyers
sedan_buyers_count = len(salaried_df[salaried_df['Make'] == 'sedan'])

# 3. Count Total Salaried Buyers
total_salaried_count = len(salaried_df)

# 4. Calculate Likelihood with division by zero check
if total_salaried_count > 0:
    likelihood = sedan_buyers_count / total_salaried_count
    print("Number of Salaried Sedan Buyers:", sedan_buyers_count)
    print("Total Number of Salaried Buyers:", total_salaried_count)
    print("Likelihood of a Salaried Person Buying a Sedan:", likelihood)
else:
    print("No salaried individuals found, cannot calculate likelihood.")
```

```python
# What evidence or data supports sheldon cooper's claim that a salaried male is an easier target for a SUV sale over a sedan sale?
```

```python
# Filter for salaried males
salaried_males = df[(df['Gender'] == 'Male') & (df['Profession'] == 'salaried')]
```

```python
# Count SUVs and Sedans
suv_count = len(salaried_males[salaried_males['Make'] == 'suv'])
sedan_count = len(salaried_males[salaried_males['Make'] == 'sedan'])
```

```python
# Calculate proportions
total_salaried_males = len(salaried_males)
```

```python
if total_salaried_males > 0:
    suv_proportion = suv_count / total_salaried_males
    sedan_proportion = sedan_count / total_salaried_males

    print(f"Proportion of SUVs: {suv_proportion:.3f}")
    print(f"Proportion of Sedans: {sedan_proportion:.3f}")

    # Simple comparison
    if suv_proportion > sedan_proportion:
        print("Salaried males prefer SUVs over sedans.")
    elif sedan_proportion > suv_proportion:
        print("Salaried males prefer sedans over SUVs.")
    else:
        print("Salaried males have equal preference for SUVs and sedans.")
else:
    print("No salaried male data found.")
```

```python
if total_salaried_males > 0:
    data = {'Car Type': ['SUV', 'Sedan'], 'Proportion': [suv_proportion, sedan_proportion]}
    df_plot = pd.DataFrame(data)

    # Using Seaborn
    sns.barplot(x='Car Type', y='Proportion', data=df_plot)
    plt.title('SUV vs. Sedan Preference (Salaried Males)')
    plt.show()

else:
    print("No salaried male data found.")
```

```python
if total_salaried_males > 0:
    labels = ['SUV', 'Sedan']
    sizes = [suv_proportion, sedan_proportion]

    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.title('SUV vs. Sedan Distribution (Salaried Males)')
    plt.show()
else:
    print("No salaried male data found.")
```

```python
if total_salaried_males > 0:
    car_types = salaried_males['Make']
    sns.countplot(x=car_types)
    plt.title('SUV vs. Sedan Counts (Salaried Males)')
    plt.show()
else:
    print("No salaried male data found.")
```

```python
# How does having a working partner influence the purchase of higher priced cars
```

```python
# Clean data
df = df.dropna(subset=['Price']) #remove rows with missing price values.

# Grouped statistics
print(df.groupby('Partner_working')['Price'].describe())
```

```python
# Box plot
sns.boxplot(x='Partner_working', y='Price', data=df)
plt.title('Automobile Spending by Partner Working Status')
plt.show()
```

```python
# Bar chart (mean comparison)
mean_prices = df.groupby('Partner_working')['Price'].mean()
mean_prices.plot(kind='bar')
plt.title('Average Automobile Spending by Partner Working Status')
plt.ylabel('Average Price')
plt.show()

# High Price analysis
high_price_threshold = df['Price'].quantile(0.75) #example threshold.
df['High Price'] = df['Price'].apply(lambda x: 1 if x > high_price_threshold else 0)
high_price_proportions = df.groupby('Partner_working')['High Price'].mean()
print(high_price_proportions)
high_price_proportions.plot(kind='bar')
plt.title('Proportion of High Price Purchases by Partner Working Status')
plt.ylabel('Proportion')
plt.show()
```
